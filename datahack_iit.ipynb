PREPROCESSING OF CATEGORICAL AND NUMERICAL DATA

```python
cate = [cname for cname in x.columns if x[cname].dtype == "object"]
nume = [cname for cname in x.columns if x[cname].dtype in ['int64', 'float64']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_trans = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')),('scaler', StandardScaler())])
```

```python
cat_trans = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')),('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(transformers=[('num', num_trans, nume),('cat', cat_trans, cate)])
```

SPLIT THE TRAINING DATASET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.3, random_state = 42)
```

LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_Reg= LogisticRegression()
```

```python
from sklearn.multioutput import MultiOutputClassifier
```

```python
final = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', MultiOutputClassifier(Log_Reg))])
```

```python
final.fit(x_train, y_train)
```

```python
predict= final.predict_proba(x_test)
```

ROC AND AUC SCORE

```python
from sklearn.metrics import roc_auc_score
```

```python
xyz_scor = roc_auc_score(y_test['xyz_vaccine'], predict[0][:, 1])
seasonal_scor = roc_auc_score(y_test['seasonal_vaccine'], predict[1][:, 1])
print(xyz_scor)
print(seasonal_scor)
```

```python
score = (xyz_scor + seasonal_scor) / 2
print(score)
```
